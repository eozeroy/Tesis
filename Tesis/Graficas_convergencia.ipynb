```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Lee el archivo de Excel
df = pd.read_excel('Resultados//Convergencia.xlsx', sheet_name='Hoja1')

# Muestra las primeras filas para verificar los datos
print(df.head())
for encabezado in df.columns:
    print(encabezado)

plt.figure(figsize=(32, 24))
```

```python

# Imprimir gráficos de líneas para cada columna vs run
for encabezado in df.columns:
    if encabezado == "Run":
        continue 
    # Configuración de estilo y contexto de Seaborn
    sns.set(style="whitegrid")  # Puedes usar otros estilos como 'darkgrid', 'ticks', etc.
    sns.set_context("talk")  # Ajusta el tamaño de las fuentes para presentaciones
    # Crear un gráfico de líneas
    sns.lineplot(x='Run', y=encabezado, data=df, color='b', linewidth=2.5)
    plt.title('Test runs', fontsize=14)
    plt.xlabel('Run', fontsize=12)
    plt.ylabel(encabezado, fontsize=12)
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.10f}'))
    plt.yticks(fontsize=9)
    plt.xticks(range(1, 12), fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.7)  # Añadir una cuadrícula estilizada
    plt.savefig(f'grafico_run_{encabezado.replace(" ", "_")}.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
test = [(0,14), (5,15), (10,18)]

for i, j in test:
    df = pd.read_excel(f'Resultados//AGTest{i}.xlsx', sheet_name=f'Test{j}')
    # Configuración de estilo y contexto de Seaborn
    sns.set(style="whitegrid")  # Puedes usar otros estilos como 'darkgrid', 'ticks', etc.
    sns.set_context("talk")  # Ajusta el tamaño de las fuentes para presentaciones
    # Crear un gráfico de líneas
    sns.lineplot(x='Generación', y=f'Mejor Aptitud T{j}', data=df, color='b', linewidth=2.5)
    plt.title('Test runs', fontsize=14)
    plt.xlabel('Generación', fontsize=12)
    plt.ylabel(f'Mejor Aptitud T{j}', fontsize=12)
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.10f}'))
    plt.yticks(fontsize=9)
    #plt.xticks(range(1, 12), fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.7)  # Añadir una cuadrícula estilizada
    plt.savefig(f'grafico_Gen_Run{i}_Test{j}.png', dpi=300, bbox_inches='tight')
    plt.show()
```
