```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Genera datos con una distribución doble pico
np.random.seed(42)
data = np.concatenate([np.random.normal(0, 1, 1000), np.random.normal(5, 1, 1000)])

# Crea un histograma con seaborn
sns.histplot(data, kde=True, bins=50, color='skyblue', stat='density')

# Añade líneas para resaltar los picos
plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(np.median(data), color='green', linestyle='dashed', linewidth=2, label='Mediana')

# Añade etiquetas y leyenda
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.title('Distribución Doble Pico')
plt.legend()

# Muestra el gráfico
plt.show()
```
