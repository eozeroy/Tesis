```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Lee el archivo de Excel
df = pd.read_excel('Resultados//Convergencia.xlsx', sheet_name='Hoja1')

# Muestra las primeras filas para verificar los datos
print(df.head())

plt.figure(figsize=(32, 24))
```

```python

# Configuracin de estilo y contexto de Seaborn
sns.set(style="whitegrid")  # Puedes usar otros estilos como 'darkgrid', 'ticks', etc.
sns.set_context("talk")  # Ajusta el tamao de las fuentes para presentaciones
# Crear un grfico de lneas
sns.lineplot(x='Run', y='Fitness', data=df, color='b', linewidth=2.5)
plt.title('Test runs', fontsize=14)
plt.xlabel('Run', fontsize=12)
plt.ylabel('Fitness', fontsize=12)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.10f}'))
plt.yticks(fontsize=9)
plt.xticks(range(1, 12), fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)  # Aadir una cuadrcula estilizada
plt.savefig('grafico_run.png', dpi=300, bbox_inches='tight')
plt.show()
```
