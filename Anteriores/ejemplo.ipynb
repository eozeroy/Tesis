```python
import numpy as np
from scipy.stats import wasserstein_distance

# Define two probability distributions (1D arrays)
distribution1 = np.array([0.2, 0.3, 0.5])
distribution2 = np.array([0.1, 0.4, 0.5])

# Compute Wasserstein distance
w_distance = 1 / wasserstein_distance(distribution1, distribution2)

print(f"Wasserstein distance: {w_distance}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def campana_de_gauss(x, media, desviacion):
    """
    Calcula el valor de la campana de Gauss en un punto dado.

    Parameters:
    - x: Punto en el eje x donde se desea evaluar la campana.
    - media: Media de la distribución normal.
    - desviacion: Desviación estándar de la distribución normal.

    Returns:
    - Valor de la campana de Gauss en el punto x.
    """
    return norm.pdf(x, media, desviacion)

# Ejemplo de uso
# media = 0.5
# desviacion = 1

dt = 44
x1 = -dt/ 2
x2 = np.abs(x1 + 1)

dl = 5 #optimizar con AG
n = 1 #optimizar con AG

desviacion = dt / dl
media = int(dt * n)

print(f"Días: {dt}")
print(f"Media: {media}")
print(f"Desviación estandar: {desviacion}")

# Crear un rango de valores en el eje x
x_values = np.linspace(x1, x2, 44)

# Calcular los valores de la campana de Gauss para cada punto en el rango
y_values = campana_de_gauss(x_values, media, desviacion)
suma_y_values = np.sum(y_values)
ajuste = (1 - suma_y_values) / dt

for i in range(len(y_values)):
    y_values[i] += ajuste

# Graficar la campana de Gauss
plt.plot(x_values, y_values, label='Campana de Gauss')
plt.title('Distribución Normal (Campana de Gauss)')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```
